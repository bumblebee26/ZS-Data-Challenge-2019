# Exploratory Data Analysis -- EDA

```python
# import required libraries
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# plot stylings
plt.style.use('fivethirtyeight')
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
# check for null values
df.isnull().sum()
```

```python
# Filling missing values in 'shot_id_number'
df.shot_id_number=df.shot_id_number.interpolate(method='linear', limit_direction = 'forward')

# Dropping irrelevent data
df=df.drop(columns=["match_event_id","match_id", "team_id","team_name","game_season"])
```

### Hot-encoding & Treatment of Missing values 


```python
# As we can see that for game with label "Man vs other_team" (i.e) "vs" have same co-oordinates(i.e)(42.982923, -71.446094)
# So, we can assume the rest matches with label "Man @ other_team" (i.e) "@" to be "Away matches"
# Thus, we can fill the missing values of column "home/away"
df["home/away"]=df["home/away"].apply(lambda x: "HOME" if "vs" in str(x) else x)
df["home/away"]=df["home/away"].apply(lambda x: "AWAY" if "@" in str(x) else x)
```

```python
df["lat/lng"] = df["lat/lng"].apply(lambda x: "HOME" if "42.982923, -71.446094" in str(x) else x)
df["lat/lng"] = df["lat/lng"].apply(lambda x: "AWAY" if str(x)!="NaN" and str(x)!="H" else x)
df["home/away"] = df["home/away"].fillna(df["lat/lng"])
```

```python
# Droping the non-useful columns
df=df.drop(columns=["lat/lng"])
df.head()
```

## Deciding the age based on Cristiano Ronaldo's Birth date and "date_of_game"
### Since, Cristiano Ronaldo is a huge personality, we can easily find his birth date, which is 5th Feb 1985

```python
# filling the missing values using interpolation
df.date_of_game=df.date_of_game.interpolate(method='pad', limit=2)
```

```python
df['date_of_game'] = pd.to_datetime(df['date_of_game'], errors='coerce')
df['Age'] = df['date_of_game'].apply(lambda x: str(x.year-1985))
```

```python
# Droping the non-useful columns
df=df.drop(columns=["date_of_game"])
df.head()
```

```python
# Forming a single column "rem_time" which is nothing but addition of the "remaining_min" and "remaining_sec"
# We fill the missing values in both columns using the duplicate versions of those columns that are given in the data set. 
# (i.e "remaining_min.1" & "remaining_sec.1")
df["remaining_min"]=df["remaining_min"].fillna(df["remaining_min.1"])
df["remaining_sec"]=df["remaining_sec"].fillna(df["remaining_sec.1"])

df["rem_time"] = df["remaining_min"]+(df["remaining_sec"] / 60)
```

```python
# Droping the non-useful columns
df=df.drop(columns=["remaining_min.1","remaining_sec.1","remaining_min","remaining_sec"])
df.head()
```

### Filling the missing data using the duplicate values provided in data set
#### 1. distance_of_shot
#### 2. power_of_shot
#### 3. knockout_match
#### 4. type_of_shot & type_of_combined_shot

```python
df["distance_of_shot"]=df["distance_of_shot"].fillna(df["distance_of_shot.1"])
df["power_of_shot"]=df["power_of_shot"].fillna(df["power_of_shot.1"])
df["knockout_match"]=df["knockout_match"].fillna(df["knockout_match.1"])
df["shot_type"]=df["type_of_shot"].fillna(df["type_of_combined_shot"])
```

```python
# Droping the non-useful columns
df=df.drop(columns=["power_of_shot.1","distance_of_shot.1","knockout_match.1",
                    "type_of_shot","type_of_combined_shot"])
df
```

### Seperating the train/test set

But we need to extract the rows in column-"is_goal" with the value 'NaN',
We need to predict the output of our system on this extracted rows

```python
# We seperate the cleaned data into training set and test set
test_set=pd.DataFrame()
train_set=pd.DataFrame()

for i in range(0,df.shape[0]):
    if pd.isnull(df.is_goal[i])==True:
        test_set=test_set.append(df.iloc[i])   
    else:
        train_set=train_set.append(df.iloc[i])
```

```python
train_set.shape
```

```python
test_set.shape
```

## Hot-Encoding

```python
train_set = train_set.sample(frac=1, random_state=1337).reset_index(drop=True) 
train_set = pd.get_dummies(train_set)
```

```python
test_set = test_set.sample(frac=1, random_state=1337).reset_index(drop=True) 
test_set = pd.get_dummies(test_set)
```

## Preparing Training and cross-validation data

```python
train_set.head()
```

```python
dataset=train_set.values
dataset
```

```python
X=np.delete(dataset,2,axis=1)
X
```

```python
Y=dataset[:,2]
Y
```

# Data Preprocessing

```python
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
X_scale = min_max_scaler.fit_transform(X)
```

```python
X_scale
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val_and_test, Y_train, Y_val_and_test = train_test_split(X_scale, Y, test_size=0.3)
X_val, X_test, Y_val, Y_test = train_test_split(X_val_and_test, Y_val_and_test, test_size=0.5)
print(X_train.shape, X_val.shape, X_test.shape, Y_train.shape, Y_val.shape, Y_test.shape)
```

# Building and Training our Neural Network

The first thing we have to do is to set up the architecture. Let’s first think about what kind of neural network architecture we want.

         Neural network architecture that we will use for our problem
![](1.png)


In words, we want to have these layers:

* Hidden layer 1: 32 neurons, ReLU activation
* Hidden layer 2: 32 neurons, ReLU activation
* Output Layer: 1 neuron, Sigmoid activation

```python
# First Step: Setting up the Architecture
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential([    
    Dense(64, activation='relu', input_shape=(106,)),    
    Dense(64, activation='relu'),   
    Dense(1, activation='sigmoid'),
])
```

```python
# Second Step: Filling in the best numbers
model.compile(optimizer='sgd', 
              loss='binary_crossentropy', 
              metrics=['accuracy'])
```

```python
hist = model.fit(X_train, Y_train,          
                 batch_size=50, epochs=10,
                 validation_data=(X_val, Y_val))
```

```python
model.evaluate(X_test, Y_test)[1]
```

```python
import matplotlib.pyplot as plt
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Val'], loc='upper right')
plt.show()
```
